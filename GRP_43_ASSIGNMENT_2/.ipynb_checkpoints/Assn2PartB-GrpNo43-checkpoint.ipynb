### 1. Decision Trees:

#### a) How are decision trees used for regression tasks different from those used for classification?

Decision trees can be used for both classification and regression tasks, but they operate differently depending on the objective. In classification, the decision tree aims to sort data points into predefined classes. The leaves of the tree represent these classes, and the path from the root to a leaf provides a decision-making process to classify a data point.

In regression, the tree is used to predict a continuous numerical value instead of a discrete class. The leaves of the tree in this case will contain a real number, often the mean of the target values of the training samples that end up in that leaf. The decision-making process remains similar, but the criteria for making splits will differ. For example, classification trees often use metrics like Gini impurity or information gain, while regression trees might use the mean squared error.

#### b) How does a decision tree make predictions for new data points?

For new data points, the decision tree applies a series of tests starting from the root node. Each test corresponds to an attribute or feature of the data point, and the outcome of the test determines which child node to visit next. This process continues recursively until a leaf node is reached. In a classification tree, the class label of the leaf node is assigned to the data point. In a regression tree, the numerical value at the leaf node is the prediction for the data point

### 2. SVMs:

#### a) What is the criterion being optimized by SVMs?

SVMs aim to find the hyperplane that maximizes the margin between different classes in the feature space. The margin is defined as the distance between the closest points (also known as support vectors) of different classes to the hyperplane. The objective is to maximize this margin while minimizing classification error, often formulated as a convex optimization problem.


#### b) How do SVMs handle the case of non-linearly separable data? What is the 'kernel trick', and why is it used in SVMs?

In cases where the data is not linearly separable, SVMs employ what is known as the "kernel trick." The kernel trick involves mapping the original feature space into a higher-dimensional space where the data becomes linearly separable. Various types of kernels like polynomial, radial basis function (RBF), and sigmoid can be used for this purpose. The kernel trick allows the SVM to find a hyperplane in the higher-dimensional space without having to explicitly compute the coordinates in that space, making the computation more efficient.

#### c) How can SVMs be extended to handle multi-class classification problems?

SVMs are inherently binary classifiers, but they can be extended to handle multi-class problems through techniques like "one-vs-one" and "one-vs-all." In "one-vs-one," an SVM model is trained for each pair of classes, resulting in \( \frac{n(n-1)}{2} \) classifiers for \( n \) classes. For a new data point, all classifiers are evaluated, and the class that wins the most pairwise contests is the final prediction. In "one-vs-all," a separate binary classifier is trained for each class against all other classes combined. The classifier that produces the highest confidence score for a new data point determines the final class label.

By understanding these aspects of Decision Trees and SVMs, more informed choices can be made when selecting a model for specific tasks, such as the breast cancer dataset analysis. Both models have their own strengths and weaknesses, and the choice often depends on the nature of the data and the specific problem at hand.

# Breast Cancer Dataset Description

The breast cancer dataset is a comprehensive collection of features that aim to describe the characteristics of cell nuclei present in breast mass images. These features are computed from a digitized image of a fine needle aspirate (FNA) of a breast mass. The dataset is commonly used for classification tasks to distinguish between malignant and benign tumors.

## Mean Features

- **Mean Radius**: The average distance from the center to points on the perimeter of the tumor.
- **Mean Texture**: The standard deviation of gray-scale values in the cell nuclei.
- **Mean Perimeter**: The average size of the core tumor.
- **Mean Area**: The average area covered by the tumor.
- **Mean Smoothness**: The average local variation in radius lengths of the tumor.
- **Mean Compactness**: Calculated as \( \frac{{\text{Perimeter}}^2}{\text{Area}} - 1.0 \).
- **Mean Concavity**: The average severity of concave portions of the contour of the tumor.
- **Mean Concave Points**: The average number of concave portions of the contour.
- **Mean Symmetry**: The average symmetry of the cell nucleus.
- **Mean Fractal Dimension**: Calculated as \( \text{"coastline approximation"} - 1 \).

## Error Features

These features represent the standard error for the corresponding 'mean' features.

- **Radius Error**
- **Texture Error**
- **Perimeter Error**
- **Area Error**
- **Smoothness Error**
- **Compactness Error**
- **Concavity Error**
- **Concave Points Error**
- **Symmetry Error**
- **Fractal Dimension Error**

## Worst Features

These features represent the "worst" or largest mean values for the corresponding 'mean' features.

### 3b

### When working with SVM in the scikit-learn library (`sklearn`), there are several model parameters and hyperparameters that can be relevant:



### Kernel Type (`kernel`):
The kernel type you choose determines the decision boundary. Common choices are `linear`, `poly`, `rbf`, and `sigmoid`.

### Regularization (`C`):
Regularization parameter `C` is a penalty term that discourages misclassification. A smaller value of `C` creates a smoother decision surface, while a larger `C` value aims to classify all training examples correctly.

### Kernel Coefficient (`gamma`):
For `rbf`, `poly`, and `sigmoid` kernels, `gamma` is a coefficient that defines how much influence a single training example has. A low `gamma` value will give a more linear decision boundary, while a high `gamma` value will model the data more closely, potentially leading to overfitting.

### Degree (`degree`):
For `poly` kernel, `degree` specifies the polynomial degree.

### Coef0 (`coef0`):
Independent term in kernel function, it is only significant in `poly` and `sigmoid`.

### Class Weights (`class_weight`):
This can be useful if your classes are imbalanced.

### To tune these hyperparameters, you would typically use techniques such as Grid Search or Randomized Search to find the optimal values:

### Grid Search (`GridSearchCV`):
This technique exhaustively searches through a specified hyperparameter grid, evaluating all the possible combinations of hyperparameter values to find the best combination using cross-validation.

### Randomized Search (`RandomizedSearchCV`):
Unlike Grid Search, Randomized Search randomly selects hyperparameter combinations to evaluate. This can be faster and more efficient, especially when there are a large number of hyperparameters and/or possible values.

### Relevant Kernel Functions in SVM---


Kernel functions are used in SVM to transform the input data into a higher-dimensional space where it becomes easier to separate the classes using a hyperplane. The choice of kernel function can significantly impact the performance of the SVM model. Here are some commonly used kernel functions in SVM:

1. **Linear Kernel**: Used when the data is linearly separable. It is computationally less intensive.
   \[
   K(x, y) = x^T y
   \]
   
2. **Polynomial Kernel**: Useful for data that is almost linearly separable but needs some transformation.
   \[
   K(x, y) = (x^T y + c)^d
   \]
   - \(c\) is a constant, \(d\) is the degree of the polynomial.
   
3. **Radial Basis Function (RBF) Kernel**: Most commonly used and works well for non-linearly separable data.
   \[
   K(x, y) = \exp(-\gamma ||x - y||^2)
   \]
   - \(\gamma\) is a parameter that needs to be set.
   
4. **Sigmoid Kernel**: Mimics the behavior of a neural network with a sigmoid activation function.
   \[
   K(x, y) = \tanh(\alpha x^T y + c)
   \]
   - \(\alpha\) and \(c\) are constants.

### Selecting the Best Kernel

The choice of kernel depends on the nature of the data:

- If the data is linearly separable or nearly so, a linear kernel is often sufficient.
- For more complex data, RBF is usually the first choice.
- Polynomial and sigmoid kernels are less commonly used but may be appropriate for specific problems.

### Kernel Selection for Our Implementation

We selected the RBF (Radial Basis Function) kernel for our SVM-BC model for the following reasons:

1. **Complexity of Data**: The breast cancer dataset contains complex relationships between features that are not linearly separable, making RBF a suitable choice.

2. **Versatility**: The RBF kernel is versatile and can perform well in a variety of classification tasks, including those where the data has non-linear boundaries.

## Model Training

From our dataset, let's create the target and predictor matrix.

- **"y"**: This is the feature we are trying to predict (Output). In this case, we are trying to predict whether our "target" is Cancer (Malignant) or not (Benign). We will use the "target" feature for this.

- **"X"**: These are the predictors, which include the remaining columns such as mean radius, mean texture, mean perimeter, mean area, mean smoothness, etc.


```python
X = df_cancer.drop(['target'], axis = 1) # We drop our "target" feature and use all the remaining features in our dataframe to train the model.
X.head()
```

## Create the Training and Testing Data

Now that we've assigned values to our "X" and "y", the next step is to import the Python library that will help us split our dataset into training and testing data.

## Parameters in SVC Function

- **C**: Regularization parameter. Controls the trade-off between maximizing the margin and minimizing the classification error.
  
- **decision_function_shape='ovr'**: Specifies the strategy for multi-class classification. 'ovr' stands for One-vs-Rest.
  
- **degree**: Degree of the polynomial kernel function ('poly'). Ignored by all other kernels.
  
- **gamma**: Kernel coefficient for 'rbf', 'poly', and 'sigmoid'. Controls the shape of the decision boundary.
  
- **kernel**: Type of kernel to be used ('linear', 'poly', 'rbf', 'sigmoid').


C: This is the regularization parameter, also known as the cost parameter. It controls the trade-off between maximizing the margin and minimizing the classification error. A smaller value of C creates a wider margin but may misclassify some data points, while a larger value of C aims for a smaller margin and could potentially overfit the data.

decision_function_shape='ovr': This parameter specifies how to handle multiple classes in the target variable. 'ovr' stands for "One-vs-Rest," meaning that for each class, one classifier is trained to separate that class from all other classes.

degree: This is the degree of the polynomial kernel function ('poly'). It is ignored by all other kernels. The default value is 3.

gamma: This is the kernel coefficient for 'rbf', 'poly', and 'sigmoid'. If gamma is 'auto', then 1/n_features will be used instead. It controls the shape of the decision boundary.

kernel: Specifies the type of kernel to be used in the algorithm. It can be 'linear', 'poly', 'rbf', 'sigmoid', or a custom kernel. The default is 'rbf'.
